```python
# A list that contains element in degrees
degree_list = [180, 90, 45, 30]
# A list that contains elemenr in radians
radian_list = list(map(math.radians, degree_list))
```

```python
# Using map method to apply math.sin method to every element inside the list
sine_list = list(map(math.sin, radian_list))
```

```python
# Using map method to apply math.cos method to every element inside the list
cosine_list = list(map(math.cos, radian_list))
```

```python
print(sine_list)
```

```python
print(cosine_list)
```

```python
# NOTE: 
# It's normal for list elements to be not completely identical to correct sine/cosine values due to rounding errors
# If you want to round them use round() function - e.g. round(0.4999999991, 5) = 0.5 
```

```python
sample_list = list(range(101))
```

```python
# print last 25 values of the list
# The list[-n] syntax gets the nth-to-last element. 
# So list[-1] gets the last element, list[-2] gets the second to last.
print(sample_list[-25:])
```

```python
# print values that fall into 1/4 to 3/4 of the list's length
# We choose the starting element 25 and ending element as 74 so that the output contains 50 elements.
print(sample_list[25:75])
```

```python
# print values that correspond to every even index
# Hint: see 2.6 indexing list section in Lab1_Lecture_Examples.ipynb
# print value that starts with index 0 and jump 2 elements and print value. For example, after printing the 
# value at index0 which is 0, it jumps two elements: from 0 to 1 and from 1 to 2. It then print value2.

print(sample_list[0::2])
```

```python
# print values that correspond to every odd index
# Hint: see 2.6 indexing list section in Lab1_Lecture_Examples.ipynb
# print value that starts with index 1 and jump 2 elements and print value. For example, after printing the 
# value at index1 which is 1, it jumps two elements: from 1 to 2 and from 2 to 3. It then print value2.

print(sample_list[1::2])
```
